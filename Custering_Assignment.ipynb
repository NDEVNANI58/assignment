```python
import numpy as np
from sklearn import cluster, datasets
from scipy import misc
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data=misc.face(gray=True)
data
```

```python
plt.gray()
plt.imshow(data)
plt.tight_layout()
```

```python
data.shape
```

```python
x=data.reshape(-1,1)
print('The length of x is %d'%len(x) , ' and dimension is %d'%x.ndim )
x
```

```python
km= KMeans(n_clusters=5)
km.fit(x)
val=km.cluster_centers_
labels=km.labels_
print('The total no of labels are: ',len(labels))
labels
```

```python
face_comp=np.choose(labels,val)
face_comp.shape = data.shape
```

```python
face_comp.shape
```

```python
vmin=data.min()
vmax= data.max()
```

```python
plt.figure(figsize=(5,5))
plt.gray()
plt.imshow(face_comp,vmin=vmin,vmax=vmax)
plt.tight_layout()
```
