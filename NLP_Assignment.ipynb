```python
from bs4 import BeautifulSoup
import urllib.request
import nltk
import pandas as pd
import re
import numpy
import string
```

```python
response = urllib.request.urlopen('http://php.net/')
html = response.read()
soup = BeautifulSoup(html,"html5lib")
```

```python
data= soup.get_text()
data
```

```python
word = nltk.word_tokenize(data)
sorted(word)
```

```python
cor=[]
for i in word:
    words= re.sub('[^a-zA-Z]','',i)
    words= words.lower()
    cor.append(words)
print(cor)
```

```python
sp=''
words=[i for i in cor if i != sp and len(i)>1 and i != string.punctuation]
sorted(words)
```

```python
z=dict(nltk.FreqDist(words))
z
```
